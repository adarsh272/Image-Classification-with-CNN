#### Handling corrupt images

Since we downloaded data from internet, we might face issues with images like wrong or unsupported file extensions or corrupt images. We should handle such images in order to prepare a smooth model

```python
import cv2
import imghdr
import os
```

```python
data_folder = 'data'
```

```python
img_extensions = ['jpeg', 'jpg', 'bmp', 'png']
```

```python
happy_images = os.listdir(os.path.join(data_folder, 'happy'))
sad_images = os.listdir(os.path.join(data_folder, 'sad'))
```

```python
for img_type in os.listdir(data_folder):
    for img in os.listdir(os.path.join(data_folder, img_type)):
        img_path = os.path.join(data_folder, img_type, img)
        try:
            image = cv2.imread(img_path)
            img_info = imghdr.what(img_path)
            if img_info not in img_extensions:
                os.remove(img_path)
        except Exception as e:
            print("Image is corrupted")
```

## Preprocessing Data
### Scaling
We need to scale all of our immages down between 0 and 1 instead of thier default range of 0, 255 for each pixeel in an image.
This will allow our deep learning model to learn faster since our optimizer function will be able to converge faster due to lower range of values of variables

**We need to insert this logic of scaling down our data range into our Keras data pipeline.
This allows for the transformation to happen in real-time or as our data gets loaded via the pipeline**

```python
data = data.map(lambda x,y: (x/255, y))
```

```python
data.as_numpy_iterator().next()[0].min()
```

### Splitting data into train, evaluate and test sets

```python
len(data)
```

```python
#splitting our batches that got created within our data pipeline
train_size = int(len(data) * .7)
evaluation_size = int(len(data) * .2)
test_size = int(len(data) * .1)
```

```python
#using the .take() method we can specify the amoutn fo data we need to fetch from our pipeline
train = data.take(train_size)
evaluation = data.skip(train_size).take(evaluation_size)
test = data.skip(train_size+evaluation_size).take(test_size)
```

## Deeep Learning Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
model = Sequential()
```

```python
# For the convolution operation layer we initialize 16 filters of size (3,3) and stride of 1
# We then use the max pooling layer to extract the pixels with largest values, indicating that these pixels highligth 
# the most crucial features of our image
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())

model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Flatten())

model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile('adam', loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

```python
logs_folder = 'logs'
```

***tf.keras.callbacks.TensorBoard*** is a callback provided by TensorFlow's Keras API that enables the integration of 
TensorBoard, **a powerful visualization tool**, during the training process of a Keras model.

```python
callback = tf.keras.callbacks.TensorBoard(logs_folder)
```

```python
hist = model.fit(train, epochs=20, validation_data=evaluation, callbacks=[callback])
```

### Performance

```python
import matplotlib.pyplot as plt
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator(): 
    X, y = batch
    y_pred = model.predict(X)
    pre.update_state(y, y_pred)
    re.update_state(y, y_pred)
    acc.update_state(y, y_pred)
```

```python
print(pre.result().numpy(), re.result().numpy(), acc.result().numpy())
```
